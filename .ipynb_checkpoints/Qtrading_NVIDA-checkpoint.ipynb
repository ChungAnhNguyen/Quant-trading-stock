## Technical 
Calculate common technical analysis indicators:

````python
# Calculate Simple Moving Averages (SMA)
stock_data['SMA20'] = stock_data['Close'].rolling(window=20).mean()  # 20-day Simple Moving Average
stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()  # 50-day Simple Moving Average
stock_data['SMA200'] = stock_data['Close'].rolling(window=200).mean()  # 200-day Simple Moving Average

# Calculate Relative Strength Index (RSI)
period = 14
delta = stock_data['Close'].diff()  # Returns a DataFrame with the difference between the values for each row

# Separate gains and losses
gain = delta.clip(lower=0)  # Positive changes, rest are 0. clip() restricts values within a specified range
loss = -delta.clip(upper=0)  # Absolute value of negative changes, rest are 0

# Use Wilder smoothing: alpha=1/period, min_periods=period, adjust=False
avg_gain = gain.ewm(alpha=1/period, min_periods=period, adjust=False).mean()
avg_loss = loss.ewm(alpha=1/period, min_periods=period, adjust=False).mean()

# Calculate RS and RSI
rs = avg_gain / avg_loss
rsi = 100 - (100 / (1 + rs))
stock_data['RSI14'] = rsi

# Calculate Bollinger Bands
rolling_std = stock_data['Close'].rolling(window=20).std()
stock_data['BB_middle'] = stock_data['Close'].rolling(window=20).mean()
stock_data['BB_upper'] = stock_data['BB_middle'] + (rolling_std * 2)
stock_data['BB_lower'] = stock_data['BB_middle'] - (rolling_std * 2)

# Display data with technical indicators
stock_data.tail()
````

## Simple Strategy Implementation

Implement a simple trading strategy based on moving average crossovers:

````python
# simulation dataUse Random Walk to simulate stock closing prices
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=300, freq='B')
# Generate lognormal random data and accumulate as prices
prices = np.random.lognormal(mean=0, sigma=0.02, size=len(dates)).cumprod() * 100
stock_data = pd.DataFrame({'Close': prices}, index=dates)

# Calculate short-term (20-day) and long-term (50-day) moving averages
stock_data['SMA_Fast'] = stock_data['Close'].rolling(window=20).mean()
stock_data['SMA_Slow'] = stock_data['Close'].rolling(window=50).mean()

# The original signal is generated based on the SMA crossover: the signal is 1 (buy) when the 20-day SMA is greater than the 50-day SMA, and -1 (sell) when the opposite is true.
stock_data['Signal'] = 0
stock_data.loc[stock_data['SMA_Fast'] > stock_data['SMA_Slow'], 'Signal'] = 1
stock_data.loc[stock_data['SMA_Fast'] < stock_data['SMA_Slow'], 'Signal'] = -1

# To avoid trading on the same day when the closing price is known, the signal is delayed by one day to simulate next day's open trading.
stock_data['Trade_Signal'] = stock_data['Signal'].shift(1)

# the market's daily return (percentage change based on closing price)
stock_data['Market_Return'] = stock_data['Close'].pct_change()

# Changes in positions (changes in positions when buying and selling incur transaction costs)
stock_data['Position_Change'] = stock_data['Trade_Signal'].diff().abs()

# Setting the transaction cost percentage (e.g. 0.1% per transaction)
transaction_cost = 0.001

#the strategy's daily return: use the previous day's trading signals multiplied by the day's market returns, minus transaction costs due to position changes
stock_data['Strategy_Return'] = stock_data['Trade_Signal'] * stock_data['Market_Return'] - transaction_cost * stock_data['Position_Change']

# Accumulated returns with initial capital set at 1
stock_data['Cumulative_Market'] = (1 + stock_data['Market_Return'].fillna(0)).cumprod()
stock_data['Cumulative_Strategy'] = (1 + stock_data['Strategy_Return'].fillna(0)).cumprod()

# plot res
plt.figure(figsize=(16, 10))

# Subfigure 1: Closing Price, SMA and Buy/Sell Signal Markers
plt.subplot(3, 1, 1)
plt.plot(stock_data.index, stock_data['Close'], label='Closed price', alpha=0.7)
plt.plot(stock_data.index, stock_data['SMA_Fast'], label='20daily average', linestyle='--')
plt.plot(stock_data.index, stock_data['SMA_Slow'], label='50daily average', linestyle='-.')
# Changes in trade signals to mark buy and sell points (when Trade_Signal jumps from -1 to 1, diff is 2; vice versa -2)
stock_data['Trade_Change'] = stock_data['Trade_Signal'].diff()
buy_signals = stock_data[stock_data['Trade_Change'] == 2].index
sell_signals = stock_data[stock_data['Trade_Change'] == -2].index

plt.scatter(buy_signals, stock_data.loc[buy_signals, 'Close'], marker='^', color='g', s=100, label='buy signal')
plt.scatter(sell_signals, stock_data.loc[sell_signals, 'Close'], marker='v', color='r', s=100, label='sell signal')
plt.title('Moving Average Crossover Strategy')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True, alpha=0.3)#

# Subfigure 2: Comparison of Cumulative Returns (Market vs Strategy)
plt.subplot(3, 1, 2)
plt.plot(stock_data.index, stock_data['Cumulative_Market'], label='Cumulative market returns')
plt.plot(stock_data.index, stock_data['Cumulative_Strategy'], label='Strategy Cumulative Returns')
plt.title('Comparison of Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative gain (initial capital = 1)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# plot position status
plt.subplot(3, 1, 3)
plt.plot(stock_data.index, stock_data['Signal'], label='position status ')
plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)

plt.fill_between(stock_data.index, stock_data['Signal'], 0, where=stock_data['Signal'] > 0, color='g', alpha=0.3, label='long')
plt.fill_between(stock_data.index, stock_data['Signal'], 0, where=stock_data['Signal'] < 0, color='r', alpha=0.3, label='short')

plt.title('Position status ')
plt.xlabel('Date')
plt.ylabel('Position')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

````

## Strategy backtesting

Calculates the return and evaluation metrics of a strategy:

````python
import numpy as np

# CLean
valid_data = stock_data.dropna()

# Calculate annualized rate of return
total_days = (valid_data.index[-1] - valid_data.index[0]).days
annual_return_strategy = (valid_data['Cumulative_Strategy'].iloc[-1] ** (365 / total_days)) - 1
annual_return_market = (valid_data['Cumulative_Market'].iloc[-1] ** (365 / total_days)) - 1

# Volatility (standard deviation)#ShuaiboHNG - QT Research
daily_vol_strategy = valid_data['Strategy_Return'].std()
annual_vol_strategy = daily_vol_strategy * np.sqrt(252)  # Assuming 252 trading days a year

daily_vol_market = valid_data['Market_Return'].std()  # Use calculated market daily returns
annual_vol_market = daily_vol_market * np.sqrt(252)

# Sharpe ratio (assuming a risk-free rate of return of 0)
sharpe_ratio_strategy = annual_return_strategy / annual_vol_strategy
sharpe_ratio_market = annual_return_market / annual_vol_market

# maximum retracement
cumulative_returns_strategy = valid_data['Cumulative_Strategy']
running_max_strategy = cumulative_returns_strategy.cummax()
drawdown_strategy = (cumulative_returns_strategy - running_max_strategy) / running_max_strategy
max_drawdown_strategy = drawdown_strategy.min()

cumulative_returns_market = valid_data['Cumulative_Market']
running_max_market = cumulative_returns_market.cummax()
drawdown_market = (cumulative_returns_market - running_max_market) / running_max_market
max_drawdown_market = drawdown_market.min()

# print res
print(f"Time period: {valid_data.index[0].date()} to {valid_data.index[-1].date()}")
print(f"Total trading days: {len(valid_data)}")
print("\n--- Strategic performance ---")

print(f"Annualized rate of return: {annual_return_strategy:.2%}")
print(f"Annualized volatility: {annual_vol_strategy:.2%}")
print(f"Sharpe ratio: {sharpe_ratio_strategy:.2f}")
print(f"maximum retracement: {max_drawdown_strategy:.2%}")
print("\n--- Benchmark (Buy & Hold) ---")
print(f"Annualized rate of return: {annual_return_market:.2%}")
print(f"Annualized volatility: {annual_vol_market:.2%}")
print(f"Sharpe ratio: {sharpe_ratio_market:.2f}")
print(f"maximum retracement: {max_drawdown_market:.2%}")
````
